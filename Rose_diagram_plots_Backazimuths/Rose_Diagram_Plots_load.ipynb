# Rose Diagram Plots of single parameters

# In this case, we made a plot of Backazimuths of seismic events recorded at a seismic station in West Africa

Station G.TAM

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Load the data from a CSV file. 
#You can change/modify to the name of your .csv file
df = pd.read_csv('TAM_BAZSLOW.csv')

# Assume the angle data is in a column named 'angle'. 
#My data has an heading in the .csv file
data = df['angle'].values

print (df)
```

```python
#Set the sizwe of the figure
plt.figure(figsize=(8,8))



# Convert to radians
data_rad = np.deg2rad(data)

# Create a histogram with 30 bins (change this as needed)
hist, bins = np.histogram(data_rad, bins=np.linspace(0.0, 2.0*np.pi, 30))


# Compute width of each bin
width = bins[1] - bins[0]

# Create a polar plot
ax = plt.subplot(1, 1, 1, polar=True)

# Create bars for each histogram bin
bars = ax.bar(bins[:len(hist)], hist, width=width,  bottom=0.0, color='red', linewidth=2, edgecolor='black')



# Set the direction of the zero angle to the top of the plot
ax.set_theta_zero_location('N')

# Set the direction of positive rotation to clockwise
ax.set_theta_direction(-1)

# Change the font size of the labels and ticks
ax.tick_params(labelsize=15)

# Set the radial labels
ax.set_xticks(np.pi/180. * np.linspace(0,  360, 8, endpoint=False))
ax.set_xticklabels(['N', 'NE', 'E',  'SE', 'S', 'SW', 'W', 'NW'], label ='bold')


#plt.colorbar(coll, orientation='horizontal')
plt.title('Backazimuth Station TAM', fontsize=15)


plt.show()

```
